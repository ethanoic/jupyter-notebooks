```python
import numpy as np
import matplotlib.pyplot as plt

# data source: https://data.gov.sg/dataset/active-cases-for-home-office-scheme
d = np.genfromtxt('data/active-cases-for-home-office-scheme.csv',
                 delimiter=',',
                 skip_header=1,
                 dtype=[('year','i4'),('no_of_cases','i4')])

# data source: https://data.gov.sg/dataset/63a59a11-f701-40e8-8b24-9c068275db69
d2 = np.genfromtxt('data/active-cases-of-renting-out-of-flat.csv',
                    delimiter=',',
                  skip_header=1,
                  dtype=[('year','i4'),('no_of_cases','i4')])

fig, ax = plt.subplots(1, figsize=(20,10))

# ax.plot(d2['year'], d2['no_of_cases'], label='RENTING')
ax.bar(d2['year'], d2['no_of_cases'], 0.15,  label='RENTING', color='green')
ax.plot(d['year'], d['no_of_cases'], label='SOHO')

ax.set_ylabel('No. of Cases')
ax.set_xlabel('Tears')
ax.legend()



plt.show()
```
